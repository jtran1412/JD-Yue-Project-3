```python
# Import libraries

import pandas as pd
from random import random, seed
```

```python
# set random seed
seed(42)
```

```python
# Set sample size (0.25%)

SAMPLE_SIZE = 0.0025
```

```python
# read and sample the CSV

df_downsized = pd.read_csv('yelp_academic_dataset_review.csv',
                 skiprows = lambda x: random() > SAMPLE_SIZE, header = None)
```

```python
# Add headers to the dataframe

headers =  ["review_id", "user_id", "business_id", "stars", "useful", "funny", "cool", "text", "date"]
df_downsized.columns = headers
```

```python
# Review the dataframe

df_downsized
```

```python
# Get the descriptive statistics of the dataframe

df_downsized.describe()
```

```python
# Convert the downsized datafrom to csv file for future use

df_downsized.to_csv('yelp.csv', index=False)
```
